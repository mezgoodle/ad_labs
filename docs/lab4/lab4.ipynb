```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Дані

```python
df = pd.read_csv('https://raw.githubusercontent.com/mezgoodle/ad_labs/master/data/telecom.csv')
df
```

# Перше завдання

```python
sns.countplot(df['churn'], palette='Greens')
```

```python
sns.barplot(x='churn', y='total day minutes', data=df,palette = "Greens");
```

```python
sns.barplot(x ='total day calls', y='voice mail plan', hue='churn', data=df, palette ='Greens');
```

# Друге завдання


```python
sns.histplot(df['total day charge']);
```

```python
x = df[(df['churn']==True)]['total day charge']
y = df[(df['churn']==False)]['total day charge']
plt.hist([x, y], color=['r','b'], alpha=0.5)
```

```python
heights_m=df[(df['churn']==True)]['total day charge']
heights_f=df[(df['churn']==False)]['total day charge']
plt.hist(heights_m,color='green',edgecolor='none',);
plt.hist(heights_f,color='blue',edgecolor='none',);
```
